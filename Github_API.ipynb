## Pagination

```python
params = { "per_page":50, "page":1}
params_2 = { "per_page":50, "page":2}
response = requests.get("https://api.github.com/users/henry0203/starred",
                       headers=headers, params=params)
response_2 = requests.get("https://api.github.com/users/henry0203",
                       headers=headers, params=params_2)
page1_repos = response.json()
print(page1_repos)
print("         ")
page2_repos = response_2.json()
print(page2_repos)
```

## Post request

```python
# payload = {"name":"test", "description":"here the repo description"} name -> repo name, description 
# requests.post("https://api.github.com/user/repos", json=payload)
```

```python
payload = {"name":"api-scraping"}
response = requests.post("https://api.github.com/user/repos", json=payload,
                        headers=headers)
status = response.status_code
print(status)
```

## Patch/ Put requests

```python
payload = {"name":"api", "description":"Super formation"}
response = requests.patch("https://api.github.com/repos/henry0203/api-scraping", 
                          json=payload, headers=headers)
status = response.status_code
print(status)
```

## Delete request

```python
response = requests.delete("https://api.github.com/repos/henry0203/api", 
                          headers=headers)
status = response.status_code
print(status)
```
